### Convert image to RGB color

### Convert to Grayscale
*  Convert image into RGB. 
   Function: cv2.cvtColor()  

    Syntax:  
    cvtColor	(	ImageObject,   
    code
    )	
    for grayscale , code= cv2.COLOR_BGR2RGB

 
* Display it using matplotlib.  
    Function: plt.title(name_of_image), plt.axis("Off"), plt.imshow(imageObject)
    *parameters for grayscale= cmap=cm.gray , vmin=0,vmax=255

```python
#your code here
#OpenCV follows BGR order, while matplotlib likely follows RGB order.
# convert image to RGB color for matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# show image with matplotlib
plt.title('RGB COLOR SPACE'),plt.axis("Off")
plt.imshow(img)
```

```python
# convert image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) #grayscale takes less processing time & memory
plt.title('GRAY COLOR SPACE'),plt.axis("Off")
plt.imshow(gray_img, cmap=cm.gray, vmin=0, vmax=255)
```

### Find Average color
Get average color of image  
Function: np.average(image, axis=0), np.uint8(average_color)

Steps:
*  Find average color per row of image
*  Find average color of average-color-per-row
*  Convert to 'uint8'
*  Print color

```python
#your code here
# find average per row
# np.average() takes in an axis argument which finds the average across that axis. 
average_color_per_row = np.average(img, axis=0)

# find average across average per row
average_color = np.average(average_color_per_row, axis=0)

# convert back to uint8
average_color = np.uint8(average_color)
print("Average Color is {}".format(average_color))
```

### Plot Average Color 
Plot average color in desired shape using matplotlib.  
Functions: np.array([[average-color]*100]*100, np.uint8), plt.title(name_of_image), plt.axis("Off"), plt.imshow(imageObject) 

Steps: 
*  Create image with average color using np.array() function
*  Plot image

```python
#your code here
# create 100 x 100 pixel array with average color value
average_color_img = np.array([[average_color]*100]*100, np.uint8)

plt.title('AVERAGE IMAGE COLOR'),plt.axis("Off")
plt.imshow(average_color_img)
```

### Plot image difference

*  Read 2 images  
   chair_1.png , chair_2.png  
   
   
*  Convert both images to grayscale


*  Plot both images
   
     
*  Find the difference between two grayscaled images.  
   Function: cv2.absdiff() 
   Syntax:  
   absdiff(firstImage,secondImage)
   
*  Plot image difference  
   code sample :- plt.imshow(img , cmap=cm.gray ,vmin=0 , vmax=255)


```python
#your code here
img1 = cv2.imread('images/chair_1.png')
img2 = cv2.imread('images/chair_2.png')

gray_img1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.title('GRAY COLOR Imgae 1'),plt.axis("Off")
plt.imshow(gray_img1, cmap=cm.gray, vmin=0, vmax=255)
```

```python
#Plot image 2
plt.title('GRAY COLOR Imgae 2'),plt.axis("Off")
plt.imshow(gray_img2, cmap=cm.gray, vmin=0, vmax=255)
```

```python
#difference between two grayscaled images
img_diff=cv2.absdiff(gray_img1,gray_img2)
plt.imshow(img_diff, cmap=cm.gray ,vmin=0 , vmax=255)
```
